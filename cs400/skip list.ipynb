```python
import random
  
class Node(object):
    '''
    Class to implement node
    '''
    def __init__(self, key, level):
        self.key = key
        self.level = level
        # list to hold references to node of different level 
        self.forward = [None]*(level+1)
    def info(self):
        return f"{self.key}, {self.level}, forward->{[i.info() if i else None for i in self.forward]}"
    def show_forward():
        return f"[{[i.key if i else None for i in self.forward]}]"
```

```python
class SkipList(object):
    '''
    Class for Skip list
    '''
    def __init__(self, max_lvl, P):
        # Maximum level for this skip list
        self.MAXLVL = max_lvl
  
        # P is the fraction of the nodes with level 
        # i references also having level i+1 references
        self.P = P
  
        # create header node and initialize key to -1
        self.header = self.createNode(self.MAXLVL, -1)
  
        # current level of skip list
        self.level = 0
        
    def info(self):
        return f"init Skpl, MAXLVL = {self.MAXLVL}, fraction = {self.P}, header = {self.header.info()}, level = {self.level}"
      
    # create new node
    def createNode(self, lvl, key):
        return Node(key, lvl)
      
    # create random level for node
    def randomLevel(self):
        lvl = 0
        r = random.random()
        # 50% to inccrement
        while r<self.P and lvl<self.MAXLVL: lvl += 1
        return lvl
  
    # Display skip list level wise
    def displayList(self):
        print("*****Skip List******")
        head = self.header
        for lvl in range(self.level+1):
            print("Level {}: ".format(lvl), end=" ")
            node = head.forward[lvl]
            while(node != None):
                print(node.key, end=" ")
                node = node.forward[lvl]
            print()
    
    # insert given key in skip list
    def insertElement(self, key):
        # create update array and initialize it
        update = [None]*(self.MAXLVL+1)
        current = self.header
        # Generate a random level for node
        rlevel = self.randomLevel()
        print(f'\n\ncurr = {current.key} | rlevel, key = {rlevel}, {key} | forward = {[i.key if i else None for i in current.forward]}')
        
  
        '''
        start from highest level of skip list
        move the current reference forward while key is greater than key of node next to current
        Otherwise inserted current in update and move one level down and continue search
        '''
        for i in range(self.level, -1, -1):
    
            # Key of next node is less than key to be inserted then we keep on moving forward on the same level
            while current.forward[i] and current.forward[i].key < key:
                print(f"curr.next={current.forward[i].key} vs key={key}")
                print(f"curr = {current.key}", end="")
                current = current.forward[i]
                print(f" -> {current.key}")
#             print(f"after loop in this lvl, curr = {current.key}")
#             print()
            
            # Key of next node is greater than the key to be inserted then 
            # we store the pointer to current node i at update[i] 
            # and move one level down and continue our search
            update[i] = current
#             print(f"insert curr to update = {[i.key if i else None for i in update]}")
            # update[i] holds the pointer to node at level i from which we moved down to level i-1 
            # and pointer of node left to insertion position at level 0
        print("end moving with level")
  
        ''' 
        reached level 0 and forward reference to right, which is desired position to insert key.
        ''' 
        
        print(f"curr = {current.key}", end="")
        current = current.forward[0]
        print(f" -> {current.key if current else current}")        
  
        '''
        if current is NULL that means we have reached to end of the level 
        or current's key is not equal to key to insert, 
        that means we have to insert node between update[0] and current node
       '''
        if current == None or current.key != key:
            '''
            If random level is greater than list's current
            level (node with highest level inserted in 
            list so far), initialize update value with reference
            to header for further use
            '''
            if rlevel > self.level:
                for i in range(self.level+1, rlevel+1):
                    update[i] = self.header
                self.level = rlevel
            # create new node with random level generated
            n = self.createNode(rlevel, key)
            print(f"new node = {n.info()}")
            for i in range(len(update)):
                if update[i]:
                    print(f"update[{i}].forward = {[j.key if j else None for j in update[i].forward]}")
                else:
                    print(f"update[{i}] = None")
            
            print("insert")
            # insert node by rearranging references 
            for i in range(rlevel+1):
                n.forward[i] = update[i].forward[i]
                update[i].forward[i] = n
#                 print(f"n.forward[{i}] = {n.forward[i]}, update[{i}].forward[{i}]={update[i].forward[i].key}")
            
            for i in range(len(update)):
                if update[i]:
                    print(f"update[{i}].forward = {[j.key if j else None for j in update[i].forward]}")
                else:
                    print(f"update[{i}] = None")
#             print(f"header = {self.header.info()}")
#             self.header.info
            self.displayList()
            print("Successfully inserted key {}\n".format(key))
  
    
```
